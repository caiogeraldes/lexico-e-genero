```python
import lexicogenero.ferramentas.diorisis_reader as dr
from lexicogenero.main import DIORISIS_PATH
import numpy as np
import pandas as pd
```

```python
corpus = dr.carrega_autores(['Herodotus', 'Thucydides', 'Appian'], DIORISIS_PATH)
```

```python
df = pd.DataFrame(columns=['sent_id', 'location', 'form', 'lemma', 'POS', 'analyses', 'id', 'file', 'author', 'text'])  
```

```python
for doc in corpus.keys():
    print(doc)
    tmp_df = dr.em_pandas(corpus, doc)
    df = pd.concat([df, tmp_df])
```

```python
from collections import Counter
Counter(df[df['POS'] == 'verb']['lemma']).most_common()
```
